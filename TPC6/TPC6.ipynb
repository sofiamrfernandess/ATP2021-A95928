```python
c=0
e=0
def criarTermo(c,e):
    c= int(input("Introduza o coefeciente que deseja:"))
    e= int(input("Introduza o expoente que deseja:"))
    return (c,e)

def criarTermo2(c, e):
    return((c, e))

def criarPol():
    return []

def inserirTermo(p,t):
    p.append(t)

def verTermo(t):
    c,e = t
    if e==0:
        return str(c)
    elif c == 1:
        return "X^"+ str(e)
    else:
        return str(c)+ "x^" + str(e)
    
def verPol(p):
    politexto="" 
    if len(p)!=0:
        politexto = politexto=verTermo(p[0])
        for i in range (1, len(p)):
            if p[i][0] >=0:        
                politexto=politexto + "+" + verTermo(p[i])
            else:
                politexto= politexto + verTermo(p[i])
        return politexto
    
def grauPol(p):
    grau = p[0][1]
    for i in range(1, len(p)):
        if p[i][1]> grau:
            grau=p[i][1]       
    return grau

def calcPol(p,x):
    valor=0
    for (c,e) in p:    
        valor= valor + c*x**e
    return valor

def derivarPol(p):
    derivada=[]
    for (c,e) in p:
        if e !=0:
            derivada.append((c*e,e-1)) 
    return derivada

def ordenarPol(p):
    simplif=[]
    simplif=p
    trocas=True
    while(trocas):
        trocas=False
        for j in range (len(simplif)-1):
            if simplif[j][1]>simplif[j+1][1]:
                simplif[j], simplif[j+1]=simplif[j+1],simplif[j]
                trocas=True   
    return simplif

def simplificarPol(simplif):
    i=0
    while (i<len(simplif)-1):
        if simplif[i][1]==simplif[i+1][1]:
            v=simplif[i][0]+simplif[i+1][0]
            t=criarTermo2(v,simplif[i][1])
            inserirTermo(simplif,t)
            simplif.pop(i)
            simplif.pop(i)
            simplif=ordenarPol(simplif)
            i=0
        else:
            i+=1
    return simplif

def tabela(p, n):
    for i in range(0,n):
        print(str(i) + " :: " + str(calcPol(p,i)))
    return tabela

```

```python
import matplotlib.pyplot as plt
def graficoPol(p):
    
    x=[n for n in range(-100,100,1)]
    y=[]
    for n in x:
          y.append(calcPol(p,n))
          
    plt.plot(x, y, color="orange", linewidth=2, marker="x", markerfacecolor="red", markersize=8)
eixoXx=[-100,100]
eixoXy=[0,0]

plt.plot(eixoXx, eixoXy, color="red", linewidth=1)

eixoYx=[0,0]
eixoYy=[-100,100]

plt.plot(eixoYx,eixoYy, color="red",linewidth=1)
          
plt.xlabel("Abcissas")
plt.ylabel("Ordenadas")
plt.title("Um gráfico que representa o polinómio:")
plt.show()
```
