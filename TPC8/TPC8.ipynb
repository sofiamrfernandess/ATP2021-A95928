````python


# BdEMD = [EMD]
# EMD = [id, nome, idade, genero, morada, modalidade, clube, email, federado, resultado]
#
# Leitura/carregamento da informação do ficheiro

def convBool(v):
    # o resultado de uma comparação é um valor boliano
    return v=='true'

def lerDataset(fnome):
    f = open(fnome, encoding="utf-8")
    bd = []
    f.readline()
    for linha in f:
        emd= []
        novaLinha= linha.strip("\n")
        campos= linha.split(",")
        emd.append("emd" + str(campos[1]))
        emd.append(campos[2])
        emd.append(campos[4]+" "+campos[3])
        emd= emd + campos[5:]  
        emd[-1] = convBool(emd[-1])#para converter para boleano???
        emd[-2] = convBool(emd[-2])
        bd.append(emd)
    return bd
    
````

````python
def modalidades(bd):
    modalidades= []
    for e in bd:
        if e[6] not in modalidades:
            modalidades.append(e[6])
    modalidades.sort()
    return modalidades
````

````python
modalidades(BD)
````

````python
def distribPorModalidade(bd):
    distribuicao={}
    for e in bd:
        if e[6] in distribuicao.keys():
            distribuicao[e[6]]= distribuicao[e[6]]+1
        else:
            distribuicao[e[6]]=1
    return distribuicao
````

````python
distribPorModalidade(BD)
````

````python
def distribPorClube(bd):
    distribuicao={}
    for e in bd:
        if e[7] in distribuicao.keys():
            distribuicao[e[7]]= distribuicao[e[7]]+1
        else:
            distribuicao[e[7]]=1
    return distribuicao
````

````python
distribPorModalidade(BD)
````

````python
def distribPorAno(bd):
    distribuicao={}
    for e in bd:
        ano= e[1][0:4] # o ano está no indice um e daí queremos ir buscar os os 4 primeiros digitos que são os do ano
        if ano in distribuicao.keys():
            distribuicao[ano]= distribuicao[ano]+1
        else:
            distribuicao[ano]=1
    return distribuicao
````

````python
distribPorAno(BD)
````

````python
def distrib(bd, index):
    distribuicao= {}
    for e in bd:
        if e[index] in distribuicao.keys():
            distribuicao[e[index]]=distribuicao[e[index]]+1
        else:
            distribuicao[e[index]]=1
    return distribuicao
````

````python
import matplotlib.pyplot as plt 
def plotDistribPorModalidade(distrib):
    modalidade = distrib.keys()
    nPessoas = distrib.values()
    plt.xlabel("Modalidades")
    plt.ylabel("Nº de pessoas")
    plt.title("Distribuição por modalidade:")
    plt.bar(modalidade,nPessoas, color="pink")
    plt.xticks(rotation=90)
    plt.show()
#plotDistribPorModalidade(distribPorModalidade(lerDataset("emd.csv")))
    
````
